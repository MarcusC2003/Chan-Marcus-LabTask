### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



**Notes**

Step 2

- This function assumes that the a child becomes an adult at 18 , adult becomes a senior at 65


```python
#1. Convert categorical values in the gender column to numeric codes.
df['Sex'] = df['Sex'].map(lambda x: 0 if x == 'male' else 1)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                #2. Create a new column that categorizes age into groups (e.g., child, adult, senior).
                                                                                                                 
display(df)
```

```python
#2. Create a new column that categorizes age into groups (e.g., child, adult, senior).

def categorize_age(x):
    if x < 18:
        return 'child'
    elif x < 65:
        return 'adult'
    else:
        return 'senior'

df['AgeGroup'] = df['Age'].map(categorize_age)
display(df)
```

Freely choose one task from each of Parts 2, 3, or 5 (3 tasks in total). Try to solve that chosen task without using Pandas, only with pure Python code. For this, convert the CSV dataset into a list (or any other native Python data type) and implement the task without using any external libraries. Then, compare the execution time of both versions (with Pandas and without it).



```python
# Read + Convert csv to list of dictionaries
import csv
import urllib.request

url = "https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"

with urllib.request.urlopen(url) as response:
    lines = [line.decode('utf-8') for line in response]

reader = csv.DictReader(lines)
records = list(reader)
```

```python
# Part 2-1. Select the names column.
names = [row['Name'] for row in records]
# display(names)
```

```python
# Part 3-1. Rows where age is above 60.
above_60 = [row for row in records if row['Age'] != '' and float(row['Age']) > 60]
# display(above_60)
```

```python
# Part 5-1. Convert Fare to Euros.
fare_eur = [float(row['Fare']) * 0.92 for row in records]
# display(fare_eur)
```

**Part 2 Step 1 Comparison**

```python
print('Pandas Time')
%timeit df['Name'].tolist()

print('\nPure Python Time')
%timeit [row['Name'] for row in records]
```

**Part 3 Step 1 Comparison**

Note: switched used .values for faster compuation

```python
print('Pandas Time')
%timeit df[df['Age'].values > 60]

print('\nPure Python Time')
%timeit [row for row in records if row['Age'] != '' and float(row['Age']) > 60]
```

**Part 5 Step 1 Comparison**

Note: switched used .values for faster compuation

```python
print('Pandas Time')
%timeit df['Fare'].values * 0.92

print('\nPure Python Time')
%timeit [float(row['Fare']) * 0.92 for row in records]
```
